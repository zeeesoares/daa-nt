```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("GnBu_d")
sns.set_style("whitegrid")
```

```python
# Load the Datasets

traffic_train = pd.read_csv('../../datasets/training_data.csv', encoding='latin1')
traffic_test = pd.read_csv('../../datasets/test_data.csv', encoding='latin1')
```

```python
# Get HOUR, MONTH, DAY OF WEEK
for df in [traffic_train, traffic_test]:
    df['record_date'] = pd.to_datetime(df['record_date'])
    df['hour'] = df['record_date'].dt.hour
    df['day_of_week'] = df['record_date'].dt.dayofweek
    df['month'] = df['record_date'].dt.month

traffic_train.head()
```

```python
# CREATE NEW FEATURES WITH TIMES

for df in [traffic_train, traffic_test]:
    is_workday = (df['day_of_week'] >= 0) & (df['day_of_week'] <= 4)
    
    is_am_peak = (df['hour'] >= 7) & (df['hour'] <= 9)
    
    is_pm_peak = (df['hour'] >= 14) & (df['hour'] <= 19)

    df["is_rush_hour"] = np.where(
        is_workday & (is_am_peak | is_pm_peak),
        1,  
        0  
    )
```
