```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("GnBu_d")
sns.set_style("whitegrid")
```

```python
# Load the Datasets

traffic_train = pd.read_csv('../../datasets/training_data.csv', encoding='latin1')
traffic_test = pd.read_csv('../../datasets/test_data.csv', encoding='latin1')
```

```python
traffic_train.describe()
```

```python
# Feature Selection

## Remove AVERAGE_PRECIPITATION due to useless information std = 0 and constant value = 0

traffic_train.drop(['AVERAGE_PRECIPITATION'], axis = 1, inplace=True)
traffic_test.drop(['AVERAGE_PRECIPITATION'], axis = 1, inplace=True)

traffic_train.head()
```

```python
## Remove city_name due to constant value = "Porto"

traffic_train.drop(['city_name'], axis = 1, inplace=True)
traffic_test.drop(['city_name'], axis = 1, inplace=True)

traffic_train.head()
```
