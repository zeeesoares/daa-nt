```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("GnBu_d")
sns.set_style("whitegrid")
```

```python
# Load the Datasets

traffic_train = pd.read_csv('../../datasets/training_data.csv', encoding='latin1')
traffic_test = pd.read_csv('../../datasets/test_data.csv', encoding='latin1')
```

```python
traffic_train.head()
```

```python
traffic_train['AVERAGE_SPEED_DIFF'].value_counts()
```

```python
traffic_train["AVERAGE_SPEED_DIFF"] = traffic_train["AVERAGE_SPEED_DIFF"].fillna("None")

speed_map = {
    'None': 0,
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Very_High': 4
}

traffic_train["AVERAGE_SPEED_DIFF"] = traffic_train["AVERAGE_SPEED_DIFF"].map(speed_map).astype(int)
traffic_train.head()
```

```python
traffic_train['LUMINOSITY'].value_counts()
```

```python
traffic_train["LUMINOSITY"] = traffic_train["LUMINOSITY"].fillna("None")
luminosity_map = {'None':0,'DARK':1,'LOW_LIGHT':2,'LIGHT':3}
traffic_train["LUMINOSITY"] = traffic_train["LUMINOSITY"].map(luminosity_map).astype(int)
traffic_train.head()
```

```python
traffic_train['AVERAGE_CLOUDINESS'].value_counts()
```

```python
traffic_train["AVERAGE_CLOUDINESS"] = traffic_train["AVERAGE_CLOUDINESS"].fillna("None")
cloud_map= {'None':0, 'algumas nuvens': 1, 'céu claro': 2, 'céu limpo': 3, 'céu pouco nublado': 4, 'nublado': 5, 'nuvens dispersas': 6, 'nuvens quebradas': 7, 'nuvens quebrados': 8, 'tempo nublado': 9}
traffic_train["AVERAGE_CLOUDINESS"] = traffic_train["AVERAGE_CLOUDINESS"].map(cloud_map).astype(int)
traffic_train.head()
```
